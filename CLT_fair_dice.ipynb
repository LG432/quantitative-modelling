```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42) # Define seed

nsides      = 6
nsamples = [5, 10, 50, 200]
nrolls      = 3000

for n in nsamples:
    sample_means = []
    sample_std   = []

    for i in range(nrolls):
        sample      = np.random.randint(1, nsides + 1, n)
        sample_means.append(np.mean(sample))

    plt.hist(sample_means, bins = 30, density = True, label = f'{n}')

    # Gaussian Overlay
    mu    = np.mean(sample_means)
    sigma = np.std(sample_means)
    x = np.linspace(np.min(sample_means), np.max(sample_means))
    plt.plot(x, norm.pdf(x, mu, sigma), '--', color = 'black')

    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
```
