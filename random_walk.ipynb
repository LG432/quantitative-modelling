```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
np.random.seed(51)

def random_walk(total_steps = 300, value_init = 1, mean = 0, std = 1):

    x = np.linspace(0, total_steps, total_steps + 1)
    
    steps = [value_init]  # initial value
    
    for i in range(total_steps):
        random = np.random.normal(loc = mean, scale = std)
        steps.append(steps[-1]*np.exp(random))  # append cumulative sum
    
    plt.plot(x, steps)


# plot multiple walks
walk_num    = 6
total_steps = 300
value_init  = 100
mean        = 0
std         = 0.05

for i in range(walk_num):
    random_walk(total_steps = total_steps, value_init = value_init, mean = mean, std = std)
    
plt.title("Random Walk")
plt.xlabel("Step")
plt.ylabel("Position")
plt.grid(True)
plt.show()
    
```
